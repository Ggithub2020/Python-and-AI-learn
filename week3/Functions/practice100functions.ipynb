## Calculator functions

```python
def add(set1,set2):
    add = set1 + set2
    print(add)
def subtract(set1,set2):
    subtract = set1 - set2
    print(subtract) 
def multiply(set1,set2):
    multiply = set1 * set2
    print(multiply)
def divide(set1,set2):
    divide = set1 / set2
    print(divide)
def modulus(set1,set2):
    modulus = set1 % set2
    print(modulus)
def exponentiation(set1,set2):
    exponentiation = set1 ** set2
    print(exponentiation)
```

```python
print(add(1, 2))
print(subtract(1, 2))
print(multiply(1, 2))
print(divide(1, 2))
print(modulus(1, 2))
print(exponentiation(1, 2))
      
```
